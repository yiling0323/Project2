```python
import random
words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]
word = random.choice(words) #randomly chose one word from words list
```

```python
def decrypt(ciphertext): #define decrypt() function, ciphertext as the parameter
    plaintest = "" #set plaintest an empty string initial
    
    for i in range(0,len(ciphertext)): #loop, from index 0 to last index in ciphertext
        if i%2 == 0: #if index number is even
            value = ord(ciphertext[i])-1 #get the value for the i index letter's ASCII TABLE value minus 1
        if i%2 == 1: #if index number is odd
            value = ord(ciphertext[i])+1 #get the value for the i index letter's ASCII TABLE value add 1
        plaintest += chr(value) #convert the value to a character, add the character to the plaintest string
            
    return plaintest 
```

```python
def hangman(): #define hangman() function
    word_letters = set(word) #seperate each letter in the randomly chosed word, put the letters in to a set
    used_letters = set() #the set for used letters
    lives = 10 #you have total 10 times to guess 
    
    #while loop for there has letter in the word that you haven't guessed yet and you still have life to guess:
    while len(word_letters) > 0 and lives > 0: 
        print("You have", lives, "lives left and you have used these letters: ", " ".join(used_letters)) #use a space to separate each used letter
        #show the letter you guessed is in the word, and "-" means the letter in the word but you haven't guessed yet:
        word_list = [letter if letter in used_letters else "-" for letter in word] 
        print("Current word: ", " ".join(word_list)) #use a space to separate letter or "-"
        user_input = input("Guess a letter: ").upper() #convert the user input letter to upper case
        print(" ") #a line between each guess
        
        if user_input not in used_letters:
            used_letters.add(user_input) #if the user input letter not in the used_letter, add it to the used_letter set
            if user_input in word_letters: #user guess right letter
                word_letters.remove(user_input)
            else: #user guess wrong letter, minus one chance/life
                print("Your letter ", user_input, " is not in word.")
                lives -= 1
        else: #user input letter already in used_letters, user guess a letter that already guessed
            print("You have alrady used that letter. Guess another letter.")
              
    if lives == 0: #user have no chance to guess, died
        print("You died, sorry. The word was", word)
        print("You'll have to guess the word correctly to decode the message")
    else: #user guess all letters in the word, call decrypt() function to get the plaintest, decode, WIN!
        print("YAY! You guessed the word", word, "!!")
        print("This was the encrypted message !")
        print("The decode message say: ", decrypt(word)) #call decrypt() function, word as argument, decode
```

```python
hangman()
```
